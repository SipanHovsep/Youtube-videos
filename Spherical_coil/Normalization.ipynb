```python
import matplotlib.pyplot as plt
import numpy as np
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

### The Basic Spherical Equations (Math Convention)
We start with the standard mathematical convention for spherical coordinates, where $\phi$ is the polar angle and $\theta$ is the azimuthal angle:

$$x = r \sin(\phi) \cos(\theta)$$
$$y = r \sin(\phi) \sin(\theta)$$
$$z = r \cos(\phi)$$

---



```python

R=4            # radius
N=5           #number of turns
points=500     #number of points
```

### Defining the Azimuthal Angle ($\theta$)
Let our parametric variable $t$ vary uniformly from 0 to 1. If we want the wire to wrap around the sphere for $N$ turns, the azimuthal angle grows linearly:

$$\theta(t) = 2\pi N t$$

**Whether the winding is uniform along the Z-axis or uniform in the polar angle depends entirely on how we define $\phi$.**

---

```python

t = np.linspace(0, 1, points)          # Parametric variable
theta = 2 * np.pi * N * t            # Azimuthal angle for N turns
```

### Method 1: Uniform Polar Angle
For a uniform polar winding, we want the polar angle $\phi$ to go from the North Pole (0) to the South Pole ($\pi$) in uniform angular steps. 

Thus, as $t$ goes from 0 to 1:
$$\phi(t) = \pi t$$

---



```python

#uniform ploar angle
phi=np.pi*t

x = R * np.sin(phi) * np.cos(theta)
y = R * np.sin(phi) * np.sin(theta)
z = R * np.cos(phi)
```

### Method 2: Uniform Z (Axial Spacing)
For uniform spacing along the vertical axis, we want $z$ to drop from $r$ to $-r$ in uniform physical steps. 
* At $t = 0$, $z = r$
* At $t = 1$, $z = -r$

Thus, the linear equation for height is:
$$z(t) = r(1 - 2t)$$

Next, we substitute this into our basic equation for $z$:
$$r \cos(\phi(t)) = r(1 - 2t)$$

Dividing both sides by $r$, we isolate the angle:
$$\cos(\phi(t)) = 1 - 2t$$

Solving for $\phi(t)$, we arrive at the final parameterization for a uniform Z winding:
$$\phi(t) = \arccos(1 - 2t)$$

```python

#uniform z 
phi_z=np.arccos(1-2*t)

x_1 = R * np.sin(phi_z) * np.cos(theta)
y_1 = R * np.sin(phi_z) * np.sin(theta)
z_1 = R * np.cos(phi_z)




fig = make_subplots(
    rows=1, cols=2,
    specs=[[{"type": "scene"}, {"type": "scene"}]],
    subplot_titles=("Uniform polar angle", "Uniform Z")
)

fig.add_trace(go.Scatter3d(x=x.flatten(), y=y.flatten(), z=z.flatten(), mode='lines'), row=1, col=1)
fig.add_trace(go.Scatter3d(x=x_1.flatten(), y=y_1.flatten(), z=z_1.flatten(), mode='lines'), row=1, col=2)

fig.update_layout(title="Parametric right Helical Winding on a Sphere")
fig.show(renderer="browser")
```
